```python
import pandas as pd
#disable false positive chained assignment warning
pd.options.mode.chained_assignment = None
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#reading in training data from 2018-2020 seasons
train_players = pd.read_csv("experimental-data/train/18-20.csv")
```

```python
#reading in test data from 2021 season
test_players = pd.read_csv("experimental-data/test/2021.csv")
```

```python
#we will only look at half PPR scoring since it is the most popular scoring system and balanced overall according to experts
train_players = train_players.drop(["StandardFantasyPoints", "PPRFantasyPoints"], axis = 1)
test_players = test_players.drop(["StandardFantasyPoints", "PPRFantasyPoints"], axis = 1)
```

```python
#sees all relevant positions
train_players.groupby('Pos').count().index
```

```python
#creating separate models for different positions
train_qb = train_players[train_players['Pos'] == 'QB']
train_rb = train_players[(train_players['Pos'] == 'RB') | (train_players['Pos'] == 'HB') \ (train_players['Pos'] == 'FB')]
train_wr = train_players[train_players['Pos'] == 'WR']
train_te = train_players[train_players['Pos'] == 'TE']

test_qb = test_players[test_players['Pos'] == 'QB']
test_rb = test_players[(test_players['Pos'] == 'RB') | (test_players['Pos'] == 'HB') \ (test_players['Pos'] == 'FB')]
test_wr = test_players[test_players['Pos'] == 'WR']
test_te = test_players[test_players['Pos'] == 'TE']
```

```python
#start by looking at qb data
train_qb
```

```python
#here we can observe which features are most strongly correlated with points earned
train_wr.corr()["HalfPPRFantasyPoints"]
```

```python
#strong correlation
sns.lmplot(x = "PassingYds", y = "HalfPPRFantasyPoints", data = train_qb, fit_reg = True, ci = None)
```

```python
train_qb.plot.hist(y = "PassingYds")
#Most qbs obtain around 200-300 passing yards per game
#notice there is an uneven distribution at the 0-50 bin, likely from backup qbs who didn't have much playing time 
```

```python
reg = LinearRegression()
```

```python
qb_predictors = ["PassingYds", "PassingTD", "PassingAtt", "Cmp", "RushingAtt", "RushingYds", "RushingTD"]
target = "HalfPPRFantasyPoints"
```

```python
reg.fit(train_qb[qb_predictors], train_qb[target])
```

```python
qb_predictions = reg.predict(test_qb[qb_predictors])
```

```python
test_qb["PredictedPoints"] = qb_predictions
test_qb
```

```python
test_qb["PredictedPoints"] = np.round(test_qb["PredictedPoints"], 2)
test_qb
```

```python
error = mean_absolute_error(test_qb["PredictedPoints"], test_qb["HalfPPRFantasyPoints"])
error
```

```python
test_qb.describe()["HalfPPRFantasyPoints"]
```

```python
qb_errors = (test_qb["PredictedPoints"] - test_qb["HalfPPRFantasyPoints"]).abs()
qb_errors
```

```python
error_by_qb = qb_errors.groupby(test_qb['Player']).mean()
error_by_qb
```

```python
points_by_qb = test_qb['HalfPPRFantasyPoints'].groupby(test_qb['Player']).mean()
```

```python
ratio_errors = error_by_qb / points_by_qb
ratio_errors
```

```python
ratio_errors.plot.hist()
```

```python
#almost 95% of predictions are within 50% of actual point values
np.percentile(ratio_errors, [2.5, 97.5])
```

```python
ratio_errors.sort_values()
```
